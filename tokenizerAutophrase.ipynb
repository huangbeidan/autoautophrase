```python
## helper function 1: check if it is a consective sublist
# list1: tokenList  list2: pattern list
def isSubList(list1, list2, s1, s2, length):
    for i in range(length):
        if(list1[s1+i] != list2[s2+i]):
            return False
    return True
```

```python
## helper function: from token to word (used for pattern mapping)
def tokenToWord(token):
    tmp = token.split(' ')[:-1]
    return ' '.join([tokendict2[t] for t in tmp] + [token.split(' ')[-1]])
```
